# Man of the match awards winners throughout the ipl journey(2008-2019).

```python

for x in range(2008,2020):
    MOM = Matches[Matches["season"] == x]
    MOM=MOM["player_of_match"].value_counts().to_dict()
    fig = plt.figure(figsize=(15,7))
    if x == x:
        fig
        plt.bar(MOM.keys(),MOM.values())
        plt.xticks(rotation=90)
        plt.xlabel("Maximum MOM awards won in year %d"%x)
        plt.show()
```

# Winning Toss is crucial!!...

```python
plt.figure(figsize=(15,7))
ax = sns.countplot(x="toss_winner",data=Matches,order = Matches['toss_winner'].value_counts().index)
for p in ax.patches:
        percentage = '{:.1f}'.format(p.get_height())
        x = p.get_x() + 0.3
        y =  p.get_height() + 0.5
        ax.annotate(percentage,(x, y))
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
fig=plt.gcf()
plt.figure(figsize=(15,10))
wins=pd.DataFrame(Matches['winner'].value_counts())
wins['name']=wins.index
plt.xticks(rotation=90,fontsize=12)
plt.bar(wins['name'],
        wins['winner'],
        color=['#15244C','#FFFF48','#292734','#EF2920','#CD202D','#ECC5F2',
               '#294A73','#D4480B','#242307','#FD511F','#158EA6','#E82865',
               '#005DB7','#C23E25','#E82865']
        ,alpha=0.8)
count=0
for i in wins['winner']:
    plt.text(count,i-5,str(i),size=15,color='Black',rotation=90)
    count+=1
plt.title('Number of matches won by each team in ipl history',fontsize=20)
plt.xlabel('Teams',fontsize=15)
plt.ylabel('Total no. of matches won(2008-2019)',fontsize=14)
plt.show()
```
